## Step 1: Import Libraries and Load Dataset

```python
import pandas as pd
import numpy as np
```

```python
# Load the dataset using tab as the separator
Data = pd.read_csv("marketing_campaign.csv",sep="\t")
```

## Step 2: Initial Inspection

```python
# View first 10 rows
Data.head(10)
```

```python
# Basic information about dataset
Data.info()
```

```python
# Summary Statistics of dataset
Data.describe()
```

## Step 3: Handling Missing Values

```python
# Check the missing values
Data.isnull().sum()
```

```python
# drop rows with missing values
# Data = Data.dropna()

         #or 

# Since the Income column has outliers, we use the median to fill the missing values
Data['Income'] = Data['Income'].fillna(Data['Income'].median())
```

```python
Data.head()
```

```python
Data.isnull().sum()
```

## Step 4: Remove Duplicates

```python
Data = Data.drop_duplicates()
```

## Step 5: Standardize Column Names

```python
# Before Standardizing Column Names
Data.columns
```

```python
Data.columns = Data.columns.str.strip().str.lower().str.replace(" ","_")
```

```python
# Before Standardizing Column Names
Data.columns
```

## Step 6: Convert Date Column

```python
# Before converting Date Column
Data['dt_customer']
```

```python
# Convert 'dt_customer' to datetime
Data['dt_customer'] = pd.to_datetime(Data['dt_customer'],format="%d-%m-%Y",dayfirst="True") 
```

```python
# After converting Date Column
Data['dt_customer']
```

## Step 7: Clean Categorical Text Data

```python
# Convert text columns to lowercase for consistency
Data['education'] = Data['education'].str.lower()
Data['marital_status'] = Data['marital_status'].str.lower()
```

```python
Data
```

```python
Data['education'].value_counts()
```

```python
Data['marital_status'].value_counts()
```

## Step 8: Check and Fix Data Types

```python
# Convert income to float if needed
Data['income'] = Data['income'].astype(float)

# Check data types
Data['income'].dtypes
```

## Step 9: Create Derived Columns

```python
# Calculate age
Data['age'] = 2025 - Data['year_birth']

# Total spent on products
Data['total_spent'] = Data[['mntwines', 'mntfruits', 'mntmeatproducts', 'mntfishproducts',
                        'mntsweetproducts', 'mntgoldprods']].sum(axis=1)
```

```python
# Export cleaned data to new CSV
Data.to_csv('customer_personality_cleaned.csv', index=False)
```
